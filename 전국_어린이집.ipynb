## groupby

! 우선 운영현황이 정상인 걸로
*   시도별& 제일 많은 지역 구별(df_c)
*   유형별(df_t)
*   보육실 면적 평균(유형별, 시도별)(df_r)
*   차량운영여부와 지역(df_car)
*   운영현황(df_m)



```python
df_c= df.groupby(['시도'])['어린이집명'].count()
df_c= df_c.reset_index()
df_c= df_c.rename(columns={'어린이집명':'수'})
df_c= df_c.sort_values(by=['수'], axis=0,ascending=False)
df_c
```

```python
# 시도별 운영현황에 따라 나누기
df_ci= df.groupby(['시도','운영현황'])['어린이집명'].count()
df_ci= df_ci.reset_index()
df_ci= df_ci.rename(columns={'어린이집명':'수'})
df_ci= df_ci.sort_values(by=['수'], axis=0,ascending=False)
df_ci
```

```python
df_t= df.groupby(['어린이집유형'])['어린이집명'].count()
df_t= df_t.reset_index()
df_t= df_t.rename(columns={'어린이집명':'수'})
df_t
```

```python
df_r=df.groupby(['어린이집유형'])['보육실면적'].mean()
df_r= df_r.reset_index()
df_r= round(df_r,2)
df_r
```

```python
df_r1=df.groupby(['시도'])['보육실면적'].mean()
df_r1= df_r1.reset_index()
df_r1= round(df_r1,2)
df_r1
```

```python
df_m=df.groupby(['운영현황'])['어린이집명'].count()
df_m= df_m.reset_index()
df_m= df_m.rename(columns={'어린이집명':'수'})
df_m
```

```python
df_car= df.groupby(['시도','통학차량운영여부'])['어린이집명'].count()
df_car= df_car.reset_index()
df_car= df_car.rename(columns={'어린이집명':'수'})
df_car= df_car.sort_values(by=['수'], axis=0,ascending=False)
df_car
```

```python
df_o= df.groupby(['인가일자'])['어린이집명'].count()
df_o= df_o.reset_index()
df_o= df_o.rename(columns={'어린이집명':'수'})
df_o= df_o.sort_values(by=['인가일자'], axis=0,ascending=False)
df_o.head()
```

```python
open= df_o["인가일자"].str.split('-')
df_o["년도"]= open.str.get(0)
```

```python
df_o= df_o.groupby(['년도'])['수'].sum()
df_o= df_o.reset_index()
df_o.head()
```

```python
df_c= df.groupby(['폐지일자'])['어린이집명'].count()
df_c= df_c.reset_index()
df_c= df_c.rename(columns={'어린이집명':'수'})
df_c= df_c.sort_values(by=['폐지일자'], axis=0,ascending=False)
df_c.head()
```

```python
close= df_c["폐지일자"].str.split('-')
df_c["년도"]= close.str.get(0)
```

```python
df_c= df_c.groupby(['년도'])['수'].sum()
df_c= df_c.reset_index()
df_c.head()
```

# 지도


```python
import folium
from folium.plugins import MarkerCluster
```

```python
df_map= df.dropna(subset=['위도'])
df_map= df_map.dropna(subset=['경도'])
df_map= df_map.reset_index()
```

```python
m_g= df_map[df_map['시도']==('경기도')]

lat= m_g['위도'].mean()
long= m_g['경도'].mean()

m= folium.Map([lat,long], zoom_start=9)
marker_cluster = MarkerCluster().add_to(m)

for i in m_g.index:
    
    sub_lat= m_g.loc[i,'위도']
    sub_long= m_g.loc[i,'경도']
    title= m_g.loc[i,'주소']
    icon_color='blue'
    
    folium.Marker(
        [sub_lat, sub_long],
        icon= folium.Icon(color=icon_color),
        popup=f'<i>{title}</i>',
        tooltip=title).add_to(marker_cluster)
    
m
```
